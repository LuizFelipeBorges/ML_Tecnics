```python
data = pd.read_csv("/content/train.csv")
X = data.drop(["Id","SalePrice"], axis=1)
y = data['SalePrice']
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
Xtr, Xval, ytr, yval = train_test_split(X,y, train_size=0.5, random_state=0)

Xtr.shape, Xval.shape, ytr.shape, yval.shape
```

```python
Xtr.head()
```

```python
cat_cols = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 
 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond',
            'Foundation',
'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'MiscFeature', 'SaleType', 
           
           'SaleCondition', 'Fence']
```

```python
from category_encoders import OneHotEncoder
cat_encoder = OneHotEncoder(cols=cat_cols, drop_invariant=True, handle_missing=0)
Xtr = cat_encoder.fit_transform(Xtr)
Xval = cat_encoder.transform(Xval)
```

```python
Xtr.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
Xtr = imputer.fit_transform(Xtr)
Xval = imputer.transform(Xval)
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
Xtr = scaler.fit_transform(Xtr)
Xval = scaler.transform(Xval)
```

```python
ytr = ytr.values
yval = yval.values
```

```python
from tensorflow import keras as tfk
```

```python
tf.random.set_seed(2)

inp = tfk.layers.Input((Xtr.shape[1], ))

hid1 = tfk.layers.Dense(100, activation='relu')(inp)

drop = tfk.layers.Dropout(0.5)(hid1)

hid2 = tfk.layers.Dense(75, activation='relu')(drop)

hid2 = tfk.layers.Dense(40, activation='relu')(hid2)

hid2 = tfk.layers.Dense(18, activation='relu')(hid2)

out = tfk.layers.Dense(1, activation='relu')(hid2)

mdl = tfk.Model(inp, out)
mdl.compile(loss='msle', optimizer='adam')
mdl.summary()
```

```python
es = tfk.callbacks.EarlyStopping(
    monitor='val_loss', min_delta=0.000001, patience=10, mode='min', restore_best_weights=True
)
```

```python
mdl.fit(Xtr, ytr, validation_data=(Xval, yval), epochs=1000, shuffle=True, batch_size=1, callbacks=[es])
```

```python
p = mdl.predict(Xval)
from sklearn.metrics import mean_squared_l_error
np.sqrt(mean_squared_log_error(yval, p))
```
